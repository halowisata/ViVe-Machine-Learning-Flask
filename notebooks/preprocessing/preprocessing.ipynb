```python
import re

# Dictionary to store the extracted key-value pairs
places = {}

# Open the text file
with open('./places.txt', 'r') as file:
    # Read each line
    for line in file:
        # Use regular expression to split the line into key and value
        matches = re.split(r'\s*:\s*', line.strip())
        
        # Check if the split was successful
        if len(matches) == 2:
            key = matches[0]
            value = matches[1].split(',')

            # Remove extra spaces from values in categories with multiple values
            if len(value) > 1:
                value = [re.sub(r'\s+', ' ', v.strip()) for v in value]

            places[key] = value
        else:
            print(f"Invalid input format: {line}")

# Print the extracted key-value pairs
print(places)
```

## Penyesuaian

```python
for index, row in df.iterrows():
    if re.search(r'Pantai', row['Place_Name']) and 'Keluarga' not in row['new_category']:
        # Append 'Keluarga' to the 'new_category' list
        df.at[index, 'new_category'].append('Keluarga')
```

```python
for index, row in df.iterrows():
    if 'Religius' in row['new_category']:
        # Find the index of 'Religius' in the list
        index_to_replace = row['new_category'].index('Religius')
        # Replace 'Religius' with 'Budaya' at the specific index
        row['new_category'][index_to_replace] = 'Budaya'
```

```python
for index, row in df.iterrows():
    if 'Edukasi' in row['new_category']:
        # Find the index of 'Religius' in the list
        index_to_replace = row['new_category'].index('Edukasi')
        # Replace 'Religius' with 'Budaya' at the specific index
        row['new_category'][index_to_replace] = 'Keluarga'
```

```python
# Define the output file path
output_file = "./output.txt"

# Open the file in write mode
with open(output_file, 'w') as file:
    # Iterate over the dictionary items
    for key, value in places.items():
        # Convert the categories list to a string
        categories = ', '.join(value)
        
        # Write the formatted string to the file
        file.write(f"{key}:{categories}\n")

# Confirm the file write
print(f"Data written to {output_file} successfully.")
```
