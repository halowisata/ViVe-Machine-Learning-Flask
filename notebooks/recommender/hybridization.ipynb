## assume we have dataset for all of the recommended items

```python
# for hybridization we use majority vote
```

```python
import pandas as pd
```

```python
place = pd.read_csv("../data/datasets/updated/tourism_with_id_updated.csv")
```

```python
final_result = place.copy()
```

```python
result_constraints_recommender = pd.read_csv("../data/output/knowledge_recommendation.csv")
result_cf_recommender = pd.read_csv("../data/output/cf_recommendation.csv")
```

```python
final_result = pd.merge(final_result, result_constraints_recommender[['Place_Id', 'score']], on='Place_Id', how='left')
final_result.rename(columns={'score': 'score_1'}, inplace=True)

final_result = pd.merge(final_result, result_cf_recommender[['Place_Id', 'score']], on='Place_Id', how='left')
final_result.rename(columns={'score': 'score_2'}, inplace=True)
```

```python
final_result['score_sum'] = final_result['score_1'].fillna(0) + final_result['score_2'].fillna(0)
```

```python
final_result.sort_values("score_sum", ascending=False, inplace=True)
```
